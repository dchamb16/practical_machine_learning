```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv('./data/mspa-survey-data.csv')
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
corr = df.loc[:, 'Personal_JavaScalaSpark':'Courses_Completed'].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, center=0, square=True, linewidths=.5, mask=mask).set_title('Heatplot of Student Preferences')
```

```python
sns.violinplot(x = 'variable', y='value', data = pd.melt(df.loc[:, 'Personal_JavaScalaSpark':'Personal_SAS'])).set_title('Personal Preference')
```

```python
sns.violinplot(x = 'variable', y='value', data = pd.melt(df.loc[:, 'Industry_JavaScalaSpark':'Industry_SAS'])).set_title('Industry Preferences')
```

```python
sns.violinplot(x = 'variable', y='value', data = pd.melt(df.loc[:, 'Professional_JavaScalaSpark':'Professional_SAS'])).set_title('Professional Preference')
```

```python
X = df[['Industry_SAS']].dropna()
```

```python
sns.distplot(X,kde=False).set_title('Industry SAS Distribution')
```

```python
sns.distplot(StandardScaler().fit_transform(X),kde=False).set_title('Industry SAS Distribution w/StandardScaler')
```

```python
sns.distplot(MinMaxScaler().fit_transform(X),kde=False).set_title('Industry SAS Distribution MinMax')
```

```python
sns.distplot(np.sqrt(X),kde=False).set_title('Industry SAS Distribution Square Root')
```
